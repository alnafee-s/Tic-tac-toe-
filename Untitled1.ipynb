```python
import math

AI = 'X'
HUMAN = 'O'
EMPTY = ' '

def print_board(board):
    for row in board:
        print(" | ".join(row))
    print("\n")

def check_winner(board):
    # Rows
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != EMPTY:
            return board[i][0]
    # Columns
    for i in range(3):
        if board[0][i] == board[1][i] == board[2][i] != EMPTY:
            return board[0][i]
    # Diagonals
    if board[0][0] == board[1][1] == board[2][2] != EMPTY:
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != EMPTY:
        return board[0][2]
    return None

def is_full(board):
    for row in board:
        if EMPTY in row:
            return False
    return True

def minimax(board, isMaximizing):
    winner = check_winner(board)
    if winner == AI:
        return 10
    if winner == HUMAN:
        return -10
    if is_full(board):
        return 0

    if isMaximizing:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = AI
                    score = minimax(board, False)
                    board[i][j] = EMPTY
                    best = max(best, score)
        return best

    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == EMPTY:
                    board[i][j] = HUMAN
                    score = minimax(board, True)
                    board[i][j] = EMPTY
                    best = min(best, score)
        return best

def find_best_move(board):
    bestScore = -math.inf
    move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if board[i][j] == EMPTY:
                board[i][j] = AI
                score = minimax(board, False)
                board[i][j] = EMPTY
                if score > bestScore:
                    bestScore = score
                    move = (i, j)

    return move


def play_game():
    board = [
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY]
    ]

    print("TIC-TAC-TOE GAME (AI = X, YOU = O)\n")

    print_board(board)

    while True:
        # Human Move
        print("Your Move:")
        r = int(input("Row (0-2): "))
        c = int(input("Col (0-2): "))
        if board[r][c] != EMPTY:
            print("Invalid Move! Try Again.\n")
            continue

        board[r][c] = HUMAN
        print_board(board)

        if check_winner(board) == HUMAN:
            print("You WIN!")
            break
        if is_full(board):
            print("DRAW!")
            break

        # AI Move
        print("AI is thinking...\n")
        r, c = find_best_move(board)
        board[r][c] = AI
        print_board(board)

        if check_winner(board) == AI:
            print("AI WINS!")
            break
        if is_full(board):
            print("DRAW!")
            break

# Run game
play_game()
```
